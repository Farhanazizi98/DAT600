```python
import time

def insertion_sort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key

def worst_case_array(n):
    return list(range(n, 0, -1))

n_values = [1000, 2000, 4000, 8000, 12000,20000]
times_python = []

print("Python results:")
for n in n_values:
    arr = worst_case_array(n)
    start = time.perf_counter()
    insertion_sort(arr)
    end = time.perf_counter()
    elapsed = end - start
    times_python.append(elapsed)
    print(f"n = {n:5d} | time = {elapsed:.4f} s")
```
